```python
import random

future_aqi = []
last_aqi = current_aqi

for i in range(5):
    change = random.randint(-8, 10)
    next_day = last_aqi + change


    next_day = max(0, next_day)

    future_aqi.append(next_day)

    last_aqi = next_day

print("Next 5 Days AQI:", future_aqi)
```

```python
import matplotlib.pyplot as plt

days = ["Day 1", "Day 2", "Day 3", "Day 4", "Day 5"]

plt.plot(days, future_aqi)
plt.xlabel("Days")
plt.ylabel("AQI")
plt.title("Future AQI Trend")
plt.show()
```
